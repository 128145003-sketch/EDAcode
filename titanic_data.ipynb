```python
import pandas as pd
df=pd.read_csv("/content/tested (1).csv")
df.head()

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data Cleaning
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    elif pd.api.types.is_numeric_dtype(df[col]):
        df[col].fillna(df[col].median(), inplace=True)
df.drop_duplicates(inplace=True)

# EDA Analysis
print("DataFrame Info:")
df.info()
print("\nDescriptive Statistics:")
print(df.describe(include='all'))
print("\nMissing values after cleaning:")
print(df.isnull().sum())

numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = df.select_dtypes(include='object').columns.tolist()

# Univariate & Outlier Identification (Box plots for numerical, Histograms for numerical, Count plots for categorical)
for col in numerical_cols:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    sns.histplot(df[col], kde=True, ax=axes[0])
    axes[0].set_title(f'Distribution of {col}')
    sns.boxplot(y=df[col], ax=axes[1])
    axes[1].set_title(f'Box plot of {col}')
    plt.tight_layout()
    plt.show()

for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=df[col], order=df[col].value_counts().index)
    plt.title(f'Count of {col}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Bivariate Plotting
if len(numerical_cols) >= 2:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df[numerical_cols[0]], y=df[numerical_cols[1]], hue=df[categorical_cols[0]] if categorical_cols else None)
    plt.title(f'Scatter Plot: {numerical_cols[0]} vs {numerical_cols[1]}')
    plt.show()

if len(numerical_cols) >= 1 and len(categorical_cols) >= 1:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[categorical_cols[0]], y=df[numerical_cols[0]])
    plt.title(f'Box Plot: {numerical_cols[0]} by {categorical_cols[0]}')
    plt.show()

if len(numerical_cols) > 1:
    plt.figure(figsize=(10, 8))
    sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Matrix of Numerical Features')
    plt.show()
```
