```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
df = pd.read_csv('/content/Smartphone_Usage_And_Addiction_Analysis_7500_Rows (1).csv')

# Initial Data Overview (EDA Analysis)
display(df.head())
df.info() # This prints to stdout, essential for initial overview
display(df.describe(include='all'))
display(df.isnull().sum())
```

```
# Data Cleaning

# Identify columns with missing values
missing_values_count = df.isnull().sum()
missing_values_percent = (missing_values_count / len(df)) * 100

# Drop columns with more than 50% missing values
threshold = 50
columns_to_drop = missing_values_percent[missing_values_percent > threshold].index
if not columns_to_drop.empty:
    df.drop(columns=columns_to_drop, inplace=True)

# Impute remaining numerical missing values with the median
for col in df.select_dtypes(include=np.number).columns:
    if df[col].isnull().any():
        df[col].fillna(df[col].median(), inplace=True)

# Impute remaining categorical missing values with the mode
for col in df.select_dtypes(include='object').columns:
    if df[col].isnull().any():
        df[col].fillna(df[col].mode()[0], inplace=True)

display(df.isnull().sum())
```

```
# Outlier Identification (Plotting)

numerical_cols = df.select_dtypes(include=np.number).columns

plt.figure(figsize=(15, 5 * (len(numerical_cols) // 3 + 1)))
for i, col in enumerate(numerical_cols):
    plt.subplot(len(numerical_cols) // 3 + 1, 3, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel('')
plt.tight_layout()
plt.show()
```

```
# Univariate Plotting

numerical_cols = df.select_dtypes(include=np.number).columns
categorical_cols = df.select_dtypes(include='object').columns

# Numerical Distributions
plt.figure(figsize=(15, 5 * (len(numerical_cols) // 3 + 1)))
for i, col in enumerate(numerical_cols):
    plt.subplot(len(numerical_cols) // 3 + 1, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Categorical Counts
plt.figure(figsize=(15, 5 * (len(categorical_cols) // 3 + 1)))
for i, col in enumerate(categorical_cols):
    plt.subplot(len(categorical_cols) // 3 + 1, 3, i + 1)
    sns.countplot(y=df[col], order=df[col].value_counts().index, palette='viridis')
    plt.title(f'Count Plot of {col}')
plt.tight_layout()
plt.show()
```

```
# Bivariate Plotting

numerical_cols = df.select_dtypes(include=np.number).columns
categorical_cols = df.select_dtypes(include='object').columns

# Numerical vs. Numerical (Pairplot for a sample)
sample_numerical_cols = numerical_cols[:min(5, len(numerical_cols))] # Limit to 5 for brevity
if len(sample_numerical_cols) > 1:
    sns.pairplot(df[sample_numerical_cols])
    plt.suptitle('Pair Plot of Sample Numerical Features', y=1.02)
    plt.show()

# Numerical vs. Categorical (Boxplots)
if len(numerical_cols) > 0 and len(categorical_cols) > 0:
    num_col_sample = numerical_cols[0]
    cat_col_sample = categorical_cols[0]
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=cat_col_sample, y=num_col_sample, data=df, palette='pastel')
    plt.title(f'{num_col_sample} Distribution by {cat_col_sample}')
    plt.xticks(rotation=45)
    plt.show()

# Categorical vs. Categorical (Heatmap)
if len(categorical_cols) > 1:
    cat_col_1 = categorical_cols[0]
    cat_col_2 = categorical_cols[1]
    plt.figure(figsize=(10, 8))
    crosstab_df = pd.crosstab(df[cat_col_1], df[cat_col_2])
    sns.heatmap(crosstab_df, annot=True, fmt='d', cmap='YlGnBu')
    plt.title(f'Crosstab of {cat_col_1} and {cat_col_2}')
    plt.show()
```
